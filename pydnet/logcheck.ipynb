```python
import numpy as np
import pandas as pd
import matplotlib
import json
import os
```

```python
logdir = 'logs/'
logfiles = os.listdir('logs')
```

```python
df = pd.DataFrame()
```

```python
for i, logfile in enumerate(logfiles):
    df[f'FPS_{i}'] = pd.read_json(logdir + logfile).drop(columns=['Time'])
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df.plot()
```
